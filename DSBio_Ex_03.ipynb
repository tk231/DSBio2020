# Integrator for the Lotka-Volterra system

Test to use SciPy's `scipy.integrate.solve_ivp` to solve for Lotka-Volterra system:

```
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

def lotkavolterra(t, z, a, b, c, d):
  x, y = z
  return [a*x - b*x*y, c*x*y - d*y]
```

Pass in the parameter values:

```
# Parameters
a, b, c, d = 2/3, 4/30, .1, 1

# Initial values
x0, y0 = 10, 15

# Time span
t_start, t_fin = 0, 50

sol = solve_ivp(fun=lotkavolterra, t_span=[t_start, t_fin], y0=[x0, y0], method='RK23', dense_output=True, rtol=1e-12, args=(a, b, c, d))
```

Compute a dense solution and plot:

```
t = np.linspace(t_start, t_fin, num=1000)
z = sol.sol(t)

plt.plot(t, z.T)
plt.xlabel('t')
plt.ylabel('Population numbers')
plt.legend(['Prey', 'Predator'])
plt.title('Lotka-Volterra System')

plt.show()
```

```
plt.plot(z[0].T, z[1].T)
plt.xlabel('Prey Numbers')
plt.ylabel('Predator numbers')
plt.legend(['Prey', 'Predator'])
plt.title('Lotka-Volterra System (Phase Space)')

plt.show()
```
